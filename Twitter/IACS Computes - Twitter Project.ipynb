```python
posWordList = []
posWords.seek(0)
for line in posWords:
    if not line[0] == ';' and len(line) > 1:
        posWordList.append(line[:-1])
```

```python
negWords = open('negative-words.txt', 'r')

negWordList = []
for line in negWords:
	#print line,
	nextWord = line.strip()
	if len(nextWord) > 0 and not nextWord.startswith(";"):
		negWordList.append(nextWord)
```
